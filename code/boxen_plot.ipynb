```python
data = pd.read_csv("..\data\salaries.csv")
salary = data.salary_in_usd
#check for nans
print("Numer of NaNs in the data = " + str(data.isnull().sum().sum()))

#Filter relevant columns
data = data[['experience_level','job_title','employment_type','salary_in_usd', 'employee_residence', 'remote_ratio', 'company_location', 'company_size']]
```

```python
#plt.figure(figsize=(200,50))
with sns.axes_style("whitegrid"):
    
    g = sns.catplot(x="experience_level", y="salary_in_usd", hue="company_size", kind="boxen",
          #  hue_order=["less than 50 employees (small)","50 to 250 employees (medium)","More than 250 employees (large)"], 
          #  order=["Entry-level","Mid-level","Senior-level","Executive-level"],
            data=data,  height=5.5, aspect=2.2
           )
    g.set(ylim=(0, 380000))

plt.savefig("experience_level.pdf")
```
